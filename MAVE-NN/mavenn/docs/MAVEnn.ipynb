# 2. Modelling Part

## 2.1 Useful Function

```python
def model_training(trainval_df,test_df,gmap_type,L):

  # Define model: L specifies the sequence length; alphabet specifies the alphabet (other choices are 'dna', 'rna', 'protein'); 
  # gpmap_type specifies the G-P map (choices are 'additive', 'neighbor', 'pairwise', 'blackbox'); regression_type specifies measurement process;
  # ge_noise_model_type specifies noise model (choices are 'Gaussian', 'Cauchy','SkewedT'); ge_heteroskedasticity_order=2 specifies noise model parameters;
  model = mavenn.Model(L=L,alphabet='protein*',gpmap_type=gmap_type,regression_type='GE',ge_noise_model_type='SkewedT',ge_heteroskedasticity_order=2)

  # Set training data
  model.set_data(x=trainval_df['sequence'],y=trainval_df['y'],validation_flags=trainval_df['validation'])

  # Train model
  model.fit(learning_rate=1e-3,epochs=500,batch_size=64,early_stopping=True,early_stopping_patience=25,verbose=False)
  
  return model
```

```python
def model_metric(model,df,set):

  # Compute variational information
  I_var, dI_var =  model.I_variational(x=df['sequence'], y=df['y'])
  print(set+ f'_I_var: {I_var:.3f} +- {dI_var:.3f} bits')

  # Compute predictive information on trainval data
  I_pred, dI_pred = model.I_predictive(x=df['sequence'], y=df['y'])
  print(set+ f'_I_pred: {I_pred:.3f} +- {dI_pred:.3f} bits')
```

```python
def model_visualization(model,df,set,wt_seq):

  # Create figure and axes
  fig, axs = plt.subplots(1, 3, figsize=[12, 4])

  # Left panel: illustrate measurement process (y vs. phi)
  ax = axs[0]

  y = df['y']

  # Compute Ï† on data
  phi = model.x_to_phi(df['sequence'])
  phi_wt_full = model.x_to_phi(wt_seq)
  yhat_wt_full = model.x_to_yhat(wt_seq)

  ## Set phi lims and create a grid in phi space
  phi_lim = [min(phi)-.5, max(phi)+.5]
  phi_grid = np.linspace(phi_lim[0], phi_lim[1], 1000)

  # Compute yhat each phi gridpoint
  yhat_grid = model.phi_to_yhat(phi_grid)

  # Compute 95% CI for each yhat
  q = [0.025, 0.975]
  yqs_grid = model.yhat_to_yq(yhat_grid, q=q)

  # Plot scatter of Ï† and y values,
  ax.scatter(phi, y,color='C0', s=5, alpha=.5, label=set+' data', rasterized=True)

  # Plot GE nonlinearity
  ax.plot(phi_grid, yhat_grid,linewidth=3, color='C1', label='$\hat{y} = g(\phi)$')

  ax.plot(phi_grid, yqs_grid[:, 0], linestyle='--', color='C1', label='95% CI')
  ax.plot(phi_grid, yqs_grid[:, 1], linestyle='--', color='C1')
  ax.axvline(phi_wt_full, color='gray', zorder=200, label='WT $\phi$')

  # Style plot
  ax.set_xlim(phi_lim)
  ax.set_xlabel('latent phenotype ($\phi$)')
  ax.set_ylabel('measurement ($y$)')
  ax.set_title('GE measurement process')
  ax.legend()

  # Center panel: illustrate model performance (y vs. yhat)
  ax = axs[1]
  yhat = model.x_to_yhat(df['sequence'])

  # Compute R^2 between yhat_test and y_test
  Rsq = np.corrcoef(yhat.ravel(), df['y'])[0, 1]**2
   
  # Plot y_test vs. yhat_test
  ax.scatter(yhat, y, color='C0', s=3, alpha=.5,
           label=set+' data')

  # Style plot
  xlim = [min(yhat), max(yhat)]
  ax.plot(xlim, xlim, '--', color='k', label='diagonal', zorder=100)
  ax.axvline(yhat_wt_full, color='gray', zorder=200)
  ax.set_xlabel('model prediction ($\hat{y}$)')
  ax.set_ylabel('measurement ($y$)')
  ax.set_title(f'Standard metric of model performance:\n$R^2$={Rsq:.3}');
  ax.legend()

  
  # Right panel: Plot model training history
  ax = axs[2]

  # Compute variational information
  I_var, dI_var =  model.I_variational(x=df['sequence'], y=df['y'])

  # Compute predictive information on trainval data
  I_pred, dI_pred = model.I_predictive(x=df['sequence'], y=df['y'])

  # Plot I_var_train, the variational information on training data as a function of epoch
  ax.plot(model.history['I_var'],label=r'I_var_train')

  # Plot I_var_val, the variational information on validation data as a function of epoch
  ax.plot(model.history['val_I_var'],label=r'I_var_val')

  # Show I_var_test, the variational information of the final model on test data
  ax.axhline(I_var, color='C2', linestyle=':',label=r'I_var_test')

  # Show I_pred_test, the predictive information of the final model on test data
  ax.axhline(I_pred, color='C3', linestyle=':',label=r'I_pred_test')

  # Style plot
  ax.set_xlabel('epochs')
  ax.set_ylabel('bits')
  ax.set_title('Training history: variational information')
  ax.set_ylim([0, 1.2*I_pred])
  ax.legend()

  # Tighten bounds on figure
  fig.tight_layout(w_pad=3)
  # fig.savefig('mpsa_ge_training.png')
  plt.show()
```

```python
def effect_map(model,model_name):

  fig, ax = plt.subplots(1,1, figsize=(12,5))

  # Retrieve G-P map parameter dict and view dict keys
  theta_dict = model.get_theta(gauge='consensus')
  theta_dict.keys()

  # Draw heatmap
  heatmap_ax, cb = mavenn.heatmap(ax=ax,
                                values=theta_dict['theta_lc'],
                                alphabet=theta_dict['alphabet'],
                                seq=model.x_stats['consensus_seq'],
                                seq_kwargs={'c':'gray', 's':25},
                                cmap='coolwarm',
                                cbar=True,
                                cmap_size='2%',
                                cmap_pad=.3,
                                ccenter=0)
  # Style heatmap (can be different between two dataset)
  #heatmap_ax.set_xticks()
  heatmap_ax.tick_params(axis='y', which='major', pad=10)
  heatmap_ax.set_xlabel('position ($l$)')
  heatmap_ax.set_ylabel('amino acid ($c$)')
  heatmap_ax.set_title(f'Additive parameters: {model_name}')

  # Style colorbar
  cb.outline.set_visible(False)
  cb.ax.tick_params(direction='in', size=20, color='white')
  cb.set_label('mutational effect ($\Delta \phi$)', labelpad=5, rotation=-90, ha='center', va='center')

  # Adjust figure and show
  fig.tight_layout(w_pad=5)
```

```python
def train_loss(model):
  fig, ax = plt.subplots(1,1,figsize=[5,5])

  # Plot loss_train, the loss computed on training data as a function of epoch
  ax.plot(model.history['loss'],
        label=r'loss_train')

  # Plot loss_val, the loss computed on validation data as a function of epoch
  ax.plot(model.history['val_loss'],
        label=r'I_var_val')

  # Style plot
  ax.set_xlabel('epochs')
  ax.set_ylabel('loss')
  ax.set_title('Training history: loss')
  ax.legend()
```

```python
def whole_model(file,gmap_type,model_name,wt):

  trainval_df, test_df = mavenn.split_dataset(file)

  L = len(file.loc[0,'sequence'])
  print(f'Sequence length: {L:d} amino acids (+ stops)')

  model=model_training(trainval_df,test_df,gmap_type,L)

  model_metric(model,trainval_df,"trainval")
  model_metric(model,test_df,"test")

  model_visualization(model,trainval_df,"trainval",wt)
  model_visualization(model,test_df,"test",wt)

  train_loss(model)

  effect_map(model,model_name)

  return model
```
