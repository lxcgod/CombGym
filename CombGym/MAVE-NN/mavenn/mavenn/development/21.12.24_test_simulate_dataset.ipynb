```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import mavenn
```

```python
# Test Model.simulate_dataset() for GE regression
model = mavenn.load_example_model("mpsa_ge_additive")
data_df = mavenn.load_example_dataset("mpsa")
sim_df = model.simulate_dataset(template_df=data_df)

plt.scatter(sim_df['phi'], sim_df['y'], alpha=.1)
```

```python
# Test Model.simulate_dataset() for MPA regression
model = mavenn.load_example_model("sortseq_full-wt_mpa_additive")
data_df = mavenn.load_example_dataset("sortseq")
sim_df = model.simulate_dataset(template_df=data_df)

# Make sure that the right number of counts were simulated for each sequence
cols = [c for c in sim_df.columns if 'ct_' in c]
np.all(sim_df[cols] == data_df[cols])

# Seems to be working!
```
