```python
import pandas as pd

df = pd.read_csv('SaCas9/SaCas9_sg3.csv')
df
```

WT序列为第一个突变体的序列

```python
#wt_seq = df.loc[df['mutant'] == 'WT', 'seq'].values[0]
mutations = []
#wt_seq
if 'mutant' not in df.columns:
    df['mutant'] = ['WT' if row['n_mut'] == 0 else '' for index, row in df.iterrows()]
first_column = df.pop('mutant')
df.insert(0, 'mutant', first_column)
wt_seq = df.loc[df['mutant'] == 'WT', 'sequence'].values[0]
wt_seq
```

突变函数

```python

def get_mutation_description(wt_seq, mutant_seq):
    descriptions = []
    for i in range(len(wt_seq)):
        if wt_seq[i] != mutant_seq[i]:
            
            descriptions.append(f"{wt_seq[i]}{i+1}{mutant_seq[i]}")
    return '-'.join(descriptions)
```

```python

for index, row in df.iterrows():
    if row['mutant'] != 'WT':
        mutant_seq = row['sequence']
        #print(row['seq'])
        mutation_desc = get_mutation_description(wt_seq, mutant_seq)
        #print(mutation_desc)
        mutations.append(mutation_desc)
        
    else:
        mutations.append('WT')
print(mutations)
```

```python
df['mutant'] = pd.Series(mutations)


df.to_csv('SaCas9/SaCas9_sg3.csv', index=False)
pd.Series(mutations)
df
```
